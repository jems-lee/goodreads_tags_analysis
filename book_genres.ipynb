```python
import pandas as pd
```

```python
book_tags = pd.read_csv("goodbooks-10k/book_tags.csv")
book = pd.read_csv("goodbooks-10k/books.csv")
ratings = pd.read_csv("goodbooks-10k/ratings.csv")
tags = pd.read_csv("goodbooks-10k/tags.csv")

book_data = pd.read_csv("toptags10/book_data.csv")
book_data_binary = pd.read_csv("toptags10/book_data_binary.csv")
```

```python
(
    book_tags
    .query("goodreads_book_id == 11084145")
    .merge(book, on="goodreads_book_id")
    .merge(tags, on="tag_id")
    .filter(items=["tag_id", "tag_name", "count"])
    .sort_values("count", ascending=False)
    .head(40)
)
```

```python
(
    book
    .filter(items=["book_id", "authors", "original_title"])
    .merge(book_data, on = "book_id")
    .query("`non-fiction` > .95")
    .head(20)
)
```

```python
(
    book
    .filter(items=["book_id", "authors", "original_title"])
    .merge(book_data, on = "book_id")
    .query("`non-fiction` > .95")
    .filter(items=["goodreads_book_id", "original_title"])
    .merge(book_tags, on="goodreads_book_id")
    .merge(tags, on="tag_id")
    .groupby(["tag_id", "tag_name"], as_index=False)
    .sum()
    .sort_values("count", ascending=False)
    .head(30)
)
```
